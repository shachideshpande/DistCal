```python
from torchuq.metric.interval import *
```

```python
predictions = torch.rand(100) 
predictions = torch.stack([predictions, predictions + torch.rand(100)], dim=1)
labels = torch.rand(100)
```

```python
plot_intervals(predictions, labels)
print(compute_coverage(predictions, labels))
print(compute_interval_length(predictions))
plot_interval_length_cdf(predictions)
```

```python
predictions[predictions > 1.0] = float('inf')
plot_intervals(predictions, labels)
print(compute_coverage(predictions, labels))
print(compute_interval_length(predictions))
plot_interval_length_cdf(predictions)
```
